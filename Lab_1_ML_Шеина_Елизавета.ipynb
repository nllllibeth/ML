```python
# задание 1 
# KMeans

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

fig1, (ax1, ax2) = plt.subplots(
    nrows = 1, ncols = 2, figsize = (8, 4))

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 200, random_state = 3, centers = 4)
ax1.scatter(X[:, 0], X[:, 1])

from sklearn.cluster import KMeans
criteries = []
for k in range(2, 10):
  KMeans_model = KMeans(n_clusters = k, random_state = 3)
  KMeans_model.fit(X)
  criteries.append(KMeans_model.inertia_)
ax2.plot(range(2, 10), criteries)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

fig = plt.Figure()

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 200, random_state = 3, centers = 4)
ax1.scatter(X[:, 0], X[:, 1])

from sklearn.cluster import KMeans
KMeans_model = KMeans(n_clusters = 4, random_state = 1)
KMeans_model.fit(X)
labels = KMeans_model.labels_
plt.scatter(X[:, 0], X[:, 1], c = labels)
```

```python
# DBSCAN 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

fig = plt.Figure()

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 200, random_state = 3, centers = 4)

from sklearn.cluster import DBSCAN 
clustering = DBSCAN(eps = 1.75, min_samples = 1).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c = clustering)
```
